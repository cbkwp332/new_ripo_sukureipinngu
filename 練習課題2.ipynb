# スクレイピングの4ステップ

**②HTMLから欲しい情報を持つタグを取得**<br>
③②で取得したタグから欲しい情報を取得<br>
④情報の保存(csvなど)

```python
from bs4 import BeautifulSoup  # BeautifulSoupライブラリをインポート
soup = BeautifulSoup(response.text)  # HTMLをパース(解析)してsoupオブジェクトを作成
```

```python
anchor_tags = soup.find_all("a")  # すべての<a>タグを取得してリストに格納
titles = []  # タイトルを格納する空のリストを作成
for anchor_tag in anchor_tags:  # 各<a>タグに対してループ
    titles.append(anchor_tag.get_text())  # タグ内のテキストをリストに追加
titles  # リストの内容を表示
```

# スクレイピングの4ステップ
①HTMLを取得<br>
②HTMLから欲しい情報を持つタグを取得<br>
**③②で取得したタグから欲しい情報を取得**<br>
④情報の保存(csvなど)

```python
DOMEIN = "https://hosh-it.github.io/scraping-practice/"  # ドメインURLを定義
anchor_tags = soup.find_all("a")  # すべての<a>タグを取得してリストに格納
anchor_tag = anchor_tags[0]  # 最初の<a>タグを取得
print(f"anchor_tag : {anchor_tag}")  # タグの内容を表示
link = anchor_tag.get("href")  # href属性(リンク先)を取得
print(f"link : {link}")  # リンクを表示
url_body = DOMEIN + link  # ドメイン+リンクで完全なURLを作成
print(f"url_body : {url_body}")  # 完全なURLを表示
response_body = requests.get(url_body)  # リンク先のページを取得
soup_body = BeautifulSoup(response_body.text)  # 取得したHTMLを解析
print(f"soup_body.find_all : \n {soup_body.find_all("p")}")  # すべての<p>タグを検索して表示
# print(f"response_body.text : \n {response_body.text}")
```

```python
body_texts = soup_body.find_all('p')  # すべての<p>タグを取得
for body_text in body_texts:  # 各<p>タグを順番に処理
    print(body_text.get_text())  # タグのテキスト部分だけを表示
```

```python
texts = ""  # 空の文字列を用意
body_texts = soup_body.find_all('p')  # すべての<p>タグを取得
for body_text in body_texts:  # 各<p>タグを順番に処理
    texts += body_text.get_text()  # テキストを連結して追加
print(texts)  # 連結したテキストを表示
```

```python
DOMEIN = "https://hosh-it.github.io/scraping-practice/"  # ドメインURLを定義
anchor_tags = soup.find_all("a")  # すべての<a>タグを取得してリストに格納
for i in range(len(anchor_tags)):
    anchor_tag = anchor_tags[i]  # 最初の<a>タグを取得
    link = anchor_tag.get("href")  # href属性(リンク先)を取得
    url_body = DOMEIN + link  # ドメイン+リンクで完全なURLを作成
    response_body = requests.get(url_body)  # リンク先のページを取得
    soup_body = BeautifulSoup(response_body.text)  # 取得したHTMLを解析
    texts = ""  # 空の文字列を用意
    body_texts = soup_body.find_all('p')  # すべての<p>タグを取得
    for body_text in body_texts:  # 各<p>タグを順番に処理
        texts += body_text.get_text()  # テキストを連結して追加
    print(texts)  # 連結したテキストを表示
```

```python
DOMEIN = "https://hosh-it.github.io/scraping-practice/"  # ドメインURLを定義
anchor_tags = soup.find_all("a")  # すべての<a>タグを取得してリストに格納
text_list = ""
for i in range(len(anchor_tags)):
    anchor_tag = anchor_tags[i]  # 最初の<a>タグを取得
    link = anchor_tag.get("href")  # href属性(リンク先)を取得
    url_body = DOMEIN + link  # ドメイン+リンクで完全なURLを作成
    response_body = requests.get(url_body)  # リンク先のページを取得
    soup_body = BeautifulSoup(response_body.text)  # 取得したHTMLを解析
    texts = ""  # 空の文字列を用意
    body_texts = soup_body.find_all('p')  # すべての<p>タグを取得
    for body_text in body_texts:  # 各<p>タグを順番に処理
        texts += body_text.get_text()  # テキストを連結して追加
    text_list += texts + "\n"
print(text_list)  # 連結したテキストを表示
```

```python
# ニュースの画像のURLを取得

img_tags = soup.find_all("img")
# print(img_tags)
img_list = []
for img_tag in img_tags:
    img_list.append(img_tag.get("src"))
print(img_list)
```

```python
import pandas as pd
news_data = {
    "記事のタイトル" :titles,  # リストの内容を表示,
    "記事の本文" :text_list,
    "画像のURL" :img_list
} 

df = pd.DataFrame(news_data)
df
```

```python
df.to_csv("fakenews.csv", index=False, encoding='utf-8-sig')
```

```python
#　ステップ１　画像データの取得
#　ステップ２　取得したデータの保存
img_url = "https://hosh-it.github.io/scraping-practice/"  # ベースURL
for j in range(len(img_list)):  # 画像リストの各画像に対してループ
    response_image = requests.get(img_url + img_list[j])  # 画像URLを構築してリクエスト
    img_name = img_list[j].split('/')[-1]  # ファイル名を取得（例：'coffee.png'）
    with open(f"img/{img_name}", "wb") as f:  # imgフォルダ内にバイナリ書き込みモードでファイルを開く
        f.write(response_image.content)  # 画像データを保存
```

```python
for index, row in df.iterrows():  # DataFrameの各行をループ処理
    img_name = row["画像のURL"]  # 画像URLの列から値を取得
    img_url_2 = DOMEIN + img_name  # ドメインと画像URLを連結して完全なURLを作成
    response_image_2 = requests.get(img_url_2)  # 画像をダウンロード
    with open(img_name, "wb") as f:  # バイナリ書き込みモードでファイルを開く
        f.write(response_image_2.content)  # 画像データをファイルに保存
```
