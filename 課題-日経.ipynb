# ⑦自動化のためにコードをまとめる

```python
import pandas as pd  # Pandas読み込み
import requests  # リクエスト送信
from bs4 import BeautifulSoup  # HTML解析

DOMAIN = "https://www.nikkei.com"  # ドメイン固定
page_start = 1  # 開始ページ
page_end = 2  # 終了ページ
rows = []  # 記事データ格納用

for page in range(page_start, page_end + 1):  # ページループ
    if page == 1:  # 1ページ目
        url = "https://www.nikkei.com/business/net-media/"  # URL指定
    else:  # 2ページ目以降
        url = f"https://www.nikkei.com/business/net-media/?page={page}"  # ページパラメータ
    response = requests.get(url)  # HTTPリクエスト
    soup = BeautifulSoup(response.text)  # HTML解析

    links = soup.find_all("a", class_="fauxBlockLink_f6t5v6i")  # リンク取得
    if page == 1:  # 1ページ目のトップ記事
        link_page_one = soup.find("a", class_="fauxBlockLink_fzas0ps")  # トップ記事取得
        if link_page_one:  # トップ記事がある場合
            links = [link_page_one] + list(links)  # リストに追加

    for link in links:  # 各リンクループ
        href = link.get("href")  # href属性取得
        title = link.get_text().replace("\u3000", "").strip()  # タイトル取得
        if not href or not title:  # 空白チェック
            continue  # スキップ

        # 本文取得
        body_link = DOMAIN + href  # URL構築
        body_response = requests.get(body_link)  # リクエスト送信
        body_soup = BeautifulSoup(body_response.text)  # HTML解析
        body_tag = body_soup.find("p", "paragraph_p18mfke4")  # 本文タグ取得
        if not body_tag:  # 本文がない場合
            continue  # スキップ
        body_text = body_tag.get_text().strip()  # テキスト抽出

        # 画像取得（本文ページから取得。imgタグが無い場合はOGP画像も確認）
        img_src = "-"  # デフォルト値
        img_tag = body_soup.find("img", class_="image_i1obkbgm")  # 画像タグ検索
        if img_tag and img_tag.get("src"):  # タグとsrc属性がある場合
            img_src = img_tag.get("src")  # URLゲット
        else:  # ない場合
            ogp = body_soup.find("meta", property="og:image")  # OGP画像検索
            if ogp and ogp.get("content"):  # OGP画像がある場合
                img_src = ogp.get("content")  # URL抽出

        rows.append({"titles": title, "body_texts": body_text, "img_urls": img_src})  # 行追加

# データフレーム化
df = pd.DataFrame(rows)  # Dataframe作成
print(f"記事数: {len(df)}")  # 記事数表示

# 表示幅調整
pd.set_option('display.width', 180)  # 表示幅設定
pd.set_option('display.max_colwidth', None)  # 列幅制限なし

styled = (  # スタイル設定
    df.style
    .set_properties(subset=['body_texts'], **{  # 本文列幅設定
        'max-width': '320px',
        'min-width': '200px',
        'white-space': 'normal'
    })
    .set_properties(subset=['img_urls'], **{  # 画像URL列幅設定
        'max-width': '320px',
        'min-width': '200px',
        'white-space': 'normal',
        'word-break': 'break-all'
    })
)

styled  # 表示
```

```python
df.to_csv("nikkei_news_2.csv", index = None, encoding= "utf-8-sig")
```

# +α画像の保存

```python
for index, row in df.iterrows():  # 各行をループ
    image_url = row["img_urls"]  # 画像URL取得
    filename = f"{index}_{row['titles']}"  # ファイル名作成
    image = requests.get(image_url)  # 画像ダウンロード
    with open("img/" + filename +".jpg", "wb") as f:  # ファイル開く
        f.write(image.content)  # 画像書き込み
```
