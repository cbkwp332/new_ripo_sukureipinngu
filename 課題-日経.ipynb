# ③HTMLの取得

```python
import requests  # HTTPリクエストを送るためのライブラリ
from bs4 import BeautifulSoup  # HTMLを解析するためのライブラリ

page = 4  # ページ番号を設定
url = "https://www.nikkei.com/business/net-media/?{page}"  # アクセスするURL
response = requests.get(url)  # URLにアクセスしてレスポンスを取得
soup = BeautifulSoup(response.text)  # HTMLをBeautifulSoupで解析
```

# ④欲しいデータを取得-【2/3】本文の取得

```python
DOMAIN = "https://www.nikkei.com"  # ドメインを変数に格納
body_link = DOMAIN + '/article/DGXZQOUC217FU0R21C25A1000000/'  # 記事のURLを作成
body_response = requests.get(body_link)  # URLにリクエストを送る
body_soup = BeautifulSoup(body_response.text)  # HTMLを解析
body_text = body_soup.find("p", "paragraph_p18mfke4").get_text()  # 段落のテキストを取得
print(body_text)  # テキストを表示
```

```python
DOMEIN = "https://www.nikkei.com"  # ドメインを変数に格納
body_texts = []

for link in links:
    body_link = link.get("href")
    body_link = DOMEIN + body_link
    body_response = requests.get(body_link)
    body_soup = BeautifulSoup(body_response.text)
    body_text = body_soup.find("p", "paragraph_p18mfke4").get_text()
    body_texts.append(body_text)
body_texts
```
