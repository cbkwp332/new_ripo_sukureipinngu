```python
import requests
url = 'https://hosh-it.github.io/scraping-practice/'
response = requests.get(url)
response.text
```

# スクレイピングの4ステップ
①HTMLを取得<br>
**②HTMLから欲しい情報を持つタグを取得**<br>
**③②で取得したタグから欲しい情報を取得**<br>
④情報の保存(csvなど)

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.text)
```

```python
# ①記事のタイトルの取得
anchor_tags = soup.find_all('a')
titles = []
for anchor_tag in anchor_tags:
    titles.append(anchor_tag.get_text())
titles
```

```python
#　②ニューズの記事の本文の取得
# ①本文が記載されているページのURL(https://hosh-it.github.io/scraping-practice/+anchorタグのhref)
# ②URLから本文の書かれたHTMLを取得
# ③②で取得したHTMLからpタグを取得
DOMAIN = 'https://hosh-it.github.io/scraping-practice/'
text_list = []
for anchor_tag in anchor_tags:
    link = anchor_tag.get('href')
    url_body = DOMAIN + link
    response_body = requests.get(url_body)
    soup_body = BeautifulSoup(response_body.text)
    body_texts = soup_body.find_all('p')
    texts = ''
    for body_text in body_texts:
        texts += body_text.get_text()
    text_list.append(texts)
text_list
```

```python
# ③ニュース画像のURL
img_tags = soup.find_all('img')
img_list = []
for img_tag in img_tags:
    img_list.append(img_tag.get('src'))
img_list
```
