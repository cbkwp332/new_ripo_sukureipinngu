```python
# â‘¡è¦ç´ ã®å±æ€§æƒ…å ±(href)
a_tags = soup.find_all('a')
a_href_list = []
for a_tag in a_tags:
    print(a_tag)
    print(a_tag.get("href"))
    a_href_list.append(a_tag.get("href"))
a_href_list
```

```python
# â‘¢è¦ç´ å†…ã®æ–‡å­—åˆ—ã®å–å¾—
a_tags = soup.find_all('a')
a_text_list = []
for a_tag in a_tags:
    print(a_tag)
    print(a_tag.get_text())
    a_text_list.append(a_tag.get_text())
a_text_list
```
